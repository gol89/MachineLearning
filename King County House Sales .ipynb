### Scaling and Train Test Split the dataset 

```python
X = df.drop('price',axis=1)
```

```python
type(X)
```

```python
X
```

```python
y = df['price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train= scaler.fit_transform(X_train)
```

```python
type(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Create the model 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()
```

```python
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(X_train,y_train,batch_size=64,epochs=400,validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses
```

```python
losses.plot()
```

```python
#Lets us evaluate the model and see
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test,predictions)
```

```python
mae
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
y_test.mean()
```

```python
#The percent of deviation from the mean is 
mae/y_test.mean()
```

```python
#thus we have a deviation of 15%
```

```python
model2 = Sequential()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model2.add(Dense(19,activation='relu'))
model2.add(Dense(50,activation='relu'))
model2.add(Dropout(0.3))
model2.add(Dense(100,activation='relu'))
model2.add(Dropout(0.4))
model2.add(Dense(250,activation='relu'))
model2.add(Dropout(0.5))
model2.add(Dense(500,activation='relu'))
model2.add(Dropout(0.5))
model2.add(Dense(250,activation='relu'))
model2.add(Dropout(0.5))
model2.add(Dense(100,activation='relu'))
model2.add(Dropout(0.4))
model2.add(Dense(50,activation='relu'))
model2.add(Dropout(0.3))
model2.add(Dense(19,activation='relu'))
model2.add(Dense(1))

model2.compile(optimizer='adam',loss='mse')
```
